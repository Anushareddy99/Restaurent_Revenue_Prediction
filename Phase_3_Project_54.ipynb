```python
import pandas as pd
import numpy as np

train = pd.read_csv("data/train.csv")
test = pd.read_csv("data/test.csv")
```

Train data

**Removing Outliers in train:**

we are removing revenue data>8M
Outlier are 8 in train data 136 rows after removing outlier it is 128 rows

```python
#remove outliers in train
index_values = train[train["revenue"]>8000000].index
train = train.drop(index_values)
train = train.set_index("Id")
train
```

picking x and y

```python
train_y = train["revenue"]# train_y is revenue
train_x = train.drop("revenue",axis=1) #droping revenue column
```


**Split data into Train and Test**:

Train is 70% and Test is 30% and shuffle train and test data.
train is feature columns and test is revenue

```python
from sklearn.model_selection import train_test_split

train_x,test_x,train_y,test_y = train_test_split(train_x,train_y,test_size = 0.3,shuffle=True,random_state=4)#split the train and test data 
#train is 70% and test is 30%
#shuffle train and test data
```

Train_x, Test_x  Datetimeindex:

date time index for year,month,dayofweek separting them and making into columns and removing open date column

```python
train_x['Year'] = pd.DatetimeIndex(train_x['Open Date']).year#year
train_x['Month'] = pd.DatetimeIndex(train_x['Open Date']).month#month
train_x['Day_of_week']=pd.DatetimeIndex(train_x['Open Date']).dayofweek# dayofweek
# drop open date
train_x=train_x.drop("Open Date",axis = 1)
train_x
#here we seperate the year,month,day for the open date column
```

```python
test_x['Year'] = pd.DatetimeIndex(test_x['Open Date']).year
test_x['Month'] = pd.DatetimeIndex(test_x['Open Date']).month
test_x['Day_of_week']=pd.DatetimeIndex(test_x['Open Date']).dayofweek
# drop open date
test_x=test_x.drop("Open Date",axis = 1)
test_x
#here we seperate the year,month,day for the open date column
```

Label Encoder and One-Hot-Encoder on train data

```python
#do feature encoding for train and test
from sklearn.preprocessing import LabelEncoder
#label encoder for city
city_label_encoder = LabelEncoder()
city_label_encoder = city_label_encoder.fit(train_x['City'])
train_x['City'] = city_label_encoder.transform(train_x['City'])
#one hot encoder for citygroup and type-train
train_x = pd.get_dummies(train_x, columns = ['City Group', 'Type'])
train_x
```

```python
le_dict = dict(zip(city_label_encoder.classes_, city_label_encoder.transform(city_label_encoder.classes_)))
le_dict #city label encoder numerical value  
```

```python
#processing step
test_x['City'] = test_x['City'].apply(lambda x: le_dict.get(x, '-1'))
test_x = pd.get_dummies(test_x, columns = ['City Group', 'Type'])
test_x
```

```python
test_x = test_x[list(train_x.columns)]#here we have train ,test equal columns for both test and train
test_x
#type dt is missing in train data
```

```python
test_x =test_x.astype({'City': 'int64'})#city is in object datatype,converting city (object)into integer
test_x.info()
```

**RMSE**: performance metrices

```python
# root mean square error(rmse)
def rmse(actual_value,predicted_value):
  actual_value=np.array(actual_value)
  predicted_value=np.array(predicted_value)
  return np.sqrt(((np.subtract(actual_value,predicted_value)**2).mean()))
```

**Linear Regression**: (base line model)

On Train_x,Train_y and y_predict the test_x without using column standardization and cross validation on data

```python
from sklearn.linear_model import LinearRegression
#baseline model
regressor = LinearRegression(n_jobs=-1)
Linear_regressor = regressor.fit(train_x,train_y)
y_pred = Linear_regressor.predict(test_x)
```

```python
rmse_lr = rmse(test_y,y_pred)
```

```python
print(f"linear regression on train rmse {rmse_lr}")
```

```python
rmse_xgbr = rmse(test_y,y_pred)
```

```python
print(f"xgboost on train rmse {rmse_xgbr}")
```

```python
from sklearn.metrics import make_scorer
my_rmse = make_scorer(rmse,greater_is_better=False)
```
